```python
import torch

N = 3
# Vận tốc của 3 đoạn đường Quận 1
F_t = torch.tensor([20, 30, 40], dtype=torch.float32)

# Cách dùng bạn hỏi: Tạo ma trận cột Fj
Fj = F_t.unsqueeze(1).expand(-1, N)
# (Lưu ý: trong forward dùng -1 cho batch, ở đây ví dụ đơn giản dùng N)

print("Ma trận Fj (mỗi cột là vận tốc 1 đường):\n", Fj)
# Kết quả:
# tensor([[20., 30., 40.],
#         [20., 30., 40.],
#         [20., 30., 40.]])
```
